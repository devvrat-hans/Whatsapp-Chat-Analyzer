```python
pattern = r"\[(\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}:\d{2}â€¯?[APM]{2})\] ([^:]+): (.*)"
```

```python
matches = re.findall(pattern, data)
```

```python
messages = []
time_info = []
senders = []
```

```python
for match in matches:
    timestamp, sender, message = match
    print(f"Timestamp: {timestamp}, Sender: {sender}, Message: {message}")
    messages.append(message)
    time_info.append(timestamp)
    senders.append(sender)
```

```python
def extract_datetime_components(timestamps):
    dates = []      
    months = []   
    years = []      
    hour_list = []  
    min_list = []  
    sec_list = [] 
    
    for timestamp in timestamps:
        date_part, time_part = timestamp.split(", ")
        
        day, month, year = map(int, date_part.split("/"))
        year = 2000 + year 
        
        time_str = time_part.split("\u202f")[0]  
        period = time_part.split("\u202f")[1]  
        hr, min, sec = map(int, time_str.split(":")) 
        
        if period == "PM" and hr != 12:
            hr += 12
        elif period == "AM" and hr == 12:
            hr = 0
            
        dates.append(day)
        months.append(month)
        years.append(year)
        hour_list.append(hr)
        min_list.append(min)
        sec_list.append(sec)
    
    return dates, months, years, hour_list, min_list, sec_list

dates, months, years, hours, minutes, seconds = extract_datetime_components(time_info)
```

```python
def convert_month_to_name(month_num):
    month_dict = {
        1: 'January',
        2: 'February', 
        3: 'March',
        4: 'April',
        5: 'May',
        6: 'June',
        7: 'July',
        8: 'August',
        9: 'September',
        10: 'October',
        11: 'November',
        12: 'December'
    }
    return month_dict[month_num]

months_list = df['month'].tolist()
month_names = [convert_month_to_name(month) for month in months_list]

df['month'] = month_names

print(df['month'].head())
```

```python
words = []

df_messages = df[df['sender'] != 'Group Notifications']

for message in df_messages['user_message']:
    words.extend(message.split())
```

```python
df_messages = df_messages[~df_messages['user_message'].str.contains('document omitted', case=False)]
df_messages = df_messages[~df_messages['user_message'].str.contains('image omitted	', case=False)]
df_messages = df_messages[~df_messages['user_message'].str.contains('audio omitted', case=False)]
df_messages = df_messages[~df_messages['user_message'].str.contains('video omitted', case=False)]
df_messages = df_messages[~df_messages['user_message'].str.contains('You deleted this message.', case=False)]
```

```python
def clean_mentions(text):
    if isinstance(text, str):
        cleaned_text = re.sub(r'@\d+\b', '', text)
        cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
        return cleaned_text
    return text

df_messages['user_message'] = df_messages['user_message'].apply(clean_mentions)
```

```python
f = open('stop_hinglish.txt', 'r')
stop_words = f.read()
print(stop_words)
```

```python
cleaned_messages = []

for message in df_messages['user_message']:
    if isinstance(message, str):
        word_list = message.split()
        filtered_words = [word for word in word_list if word.lower() not in stop_words]
        cleaned_message = ' '.join(filtered_words)
        cleaned_messages.append(cleaned_message)
    else:
        cleaned_messages.append(message)

df_messages['cleaned_message'] = cleaned_messages
```
